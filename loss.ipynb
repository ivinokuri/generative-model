```python
import scipy.stats
import numpy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = numpy.arange(0, 20)
```

```python
nu = numpy.exp(numpy.log(2))
for mu in range(1, 20):
    plt.plot(scipy.stats.poisson.pmf(x, mu=mu), color='blue', lw=3)
    plt.plot(scipy.stats.gamma.pdf(x, a=2*nu*mu, loc=-mu, scale=1/nu), color='orange')
```

```python
mu = 1
plt.figure(figsize=(20, 5))
plt.plot(scipy.stats.poisson.pmf(x, mu=mu), color='black', lw=2)
nu = 0.125
for nu in numpy.logspace(-2, 4, 7, base=2):
    plt.plot(scipy.stats.gamma.pdf(x, a=2*nu*mu, loc=-mu, scale=1/nu), label=f'nu={nu:.3f} var={2/nu*mu:.3f}')
plt.legend()
```

```python
plt.figure(figsize=(20, 5))
plt.plot(-scipy.stats.poisson.logpmf(x, mu=mu), color='black', lw=2)
nu = 0.125
for nu in numpy.logspace(-2, 4, 7, base=2):
    plt.plot(-scipy.stats.gamma.logpdf(x, a=2*nu*mu, loc=-mu, scale=1/nu), label=f'nu={nu:.3f} var={2/nu*mu:.3f}')
plt.legend()
```
