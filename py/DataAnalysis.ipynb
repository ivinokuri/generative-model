```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
from datetime import datetime

cwd = os.getcwd()
print(cwd)
```

## Loading normal data

```python
NORMAL_DATA_PATH = "../robot-data/thesis_print-master/available runs/navigation scenarios/move_base fault/normal/type1/building_normal1.xlsx"
```

```python
df = pd.read_excel(NORMAL_DATA_PATH)
```

```python
def get_columns(dataframe, columns):
    COLUMNS = []
    for c in columns:
        COLUMNS.append([col for col in dataframe.columns if c in col])
    COLUMNS = np.array(COLUMNS).T
    return COLUMNS
```

```python
CLS = ['Messages-Age', 'Mean_Consecutive', 'Counter']
COLUMN_NAMES = get_columns(df, CLS)
```

fix, axs = plt.subplots(len(COLUMN_NAMES), figsize=(15, 400))
for i, k in enumerate(COLUMN_NAMES):
    ma = df[k[0]]
    ma = ma - ma[0]
    c = df[k[2]]
    axs[i].plot(ma, c)
    axs[i].set_title(k[2])
plt.show()
