```python
import numpy
import matplotlib.pyplot as plt
```

```python
losses = numpy.genfromtxt('lines_building_counts_only_.csv', delimiter=',', skip_header=1)
```

```python
skip = 10
normals= losses[skip:,1:11]
anomals = losses[skip:, 11:]
```

```python
x = numpy.arange(len(normals))
mean = numpy.mean(normals, axis=1)
std = numpy.std(normals, axis=1)
plt.plot(mean, color='black', lw=3)
plt.fill_between(x, mean - std, mean+std, color='gray', alpha=0.5)
plt.plot(anomals)
```
