```python
import numpy as np
import os
import matplotlib.pyplot as plt
from scipy.stats import ks_2samp
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
def load_directory(scenario):
    x = np.loadtxt("./ROC/x_" + scenario,
                 delimiter=",")
    y = np.loadtxt("./ROC/y_" + scenario,
                 delimiter=",")
    return x, y
```

```python
def plot_dp_gmm(x, y, title=""):
    print('=======================================')
    print(title)
#     print(x)
#     print(y)
    fig, axs = plt.subplots(1)
    axs.set(xlim=(-0.1, 1.1), ylim=(-0.1, 1.1))
    axs.plot(x, y, marker='o')
    axs.plot(axs.get_xlim(), axs.get_ylim(), ls="--", c=".3")
    axs.set_title(title)
    axs.set_ylabel('True Positive Rate')
    axs.set_xlabel('False Positive Rate')
    plt.show()

    auroc = 0
    for i in range(len(x)-1):
        auroc +=((x[i+1]-x[i]) * (y[i+1]+y[i]))/2
    print('AUROC:', auroc)
    print('=======================================')
```

```python
TITLE_ANOMALY = 'Building with laser fault'
SCENARIO = 'ROC Building with laser fault.csv'
x, y = load_directory(SCENARIO)
plot_dp_gmm(x, y, TITLE_ANOMALY)

TITLE_ANOMALY = 'Building with software fault'
SCENARIO = 'ROC Building with software fault.csv'
x, y = load_directory(SCENARIO)
plot_dp_gmm(x, y, TITLE_ANOMALY)

TITLE_ANOMALY = 'Building with velocity attack'
SCENARIO = 'ROC Building with velocity attack.csv'
x, y = load_directory(SCENARIO)
plot_dp_gmm(x, y, TITLE_ANOMALY)

TITLE_ANOMALY = 'Corr with Laser fault'
SCENARIO = 'ROC Corr with Laser fault.csv'
x, y = load_directory(SCENARIO)
plot_dp_gmm(x, y, TITLE_ANOMALY)

TITLE_ANOMALY = 'Corr with Obstacles'
SCENARIO = 'ROC Corr with Obstacles.csv'
x, y = load_directory(SCENARIO)
plot_dp_gmm(x, y, TITLE_ANOMALY)

TITLE_ANOMALY = 'Pick and miss cup'
SCENARIO = 'ROC Pick and miss cup.csv'
x, y = load_directory(SCENARIO)
plot_dp_gmm(x, y, TITLE_ANOMALY)

TITLE_ANOMALY = 'Pick and restricted vision'
SCENARIO = 'ROC Pick and restricted vision.csv'
x, y = load_directory(SCENARIO)
plot_dp_gmm(x, y, TITLE_ANOMALY)

TITLE_ANOMALY = 'Pick and stolen'
SCENARIO = 'ROC Pick and stolen.csv'
x, y = load_directory(SCENARIO)
plot_dp_gmm(x, y, TITLE_ANOMALY)

TITLE_ANOMALY = 'Pick and stuck'
SCENARIO = 'ROC Pick and stuck.csv'
x, y = load_directory(SCENARIO)
plot_dp_gmm(x, y, TITLE_ANOMALY)

```
